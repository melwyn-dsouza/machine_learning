```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    df  = pd.read_excel("data.xlsx")
    df = df.dropna() 
    X,y = df.X, df.Y
    X = (X - np.mean(X))/np.std(X)
    linearRegression(X,y)
```

```python
def gradient_descent(b, w, alpha,X,y,max_iters):
    m = X.shape[0]
    cost = []
    for i in range(max_iters):
        #y predictions or hypothesis for linear regression
        H = (w*X)+b
        #mean squared error
        cost.append((np.square(np.sum(H-y)))/(2*m))
        #plot X,y and X,y_predicted
        plt.scatter(X,y, label = "Data")
        plt.plot(X,H,'r-', label = "Hypothesis")
        plt.title("Iteration: "+str(i))
        plt.xlabel("X")
        plt.ylabel("y")
        plt.legend()
        plt.show()
        w = w - alpha*((np.sum((H-y)*X))/m)
        b = b - alpha*((np.sum(H-y))/m)
    

    plt.plot(range(max_iters),cost, "g-")
    plt.title("Cost Function")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Cost J(X,y)")
    plt.show()
        
    return w,b
```

```python
def linearRegression(X,y):
    b = 0
    w = 0
    alpha = 0.2
    max_iters = 20
    b,w = gradient_descent(b,w,alpha,X,y,max_iters)
    
```

```python
if __name__ == '__main__':
    main()
```
